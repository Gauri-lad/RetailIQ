**Data Cleaning & Feature Engineering**

```python
# Convert date columns (DD/MM/YYYY format)

sales_df['Order Date'] = pd.to_datetime(
    sales_df['Order Date'],
    format='%d/%m/%Y'
)

sales_df['Ship Date'] = pd.to_datetime(
    sales_df['Ship Date'],
    format='%d/%m/%Y'
)

print("Date columns converted successfully!")
```

```python
sales_df[['Order Date', 'Ship Date']].head(10)
```

```python
sales_df[['Order Date', 'Ship Date']].dtypes
```

```python
sales_df['Shipping Days'] = (
    sales_df['Ship Date'] - sales_df['Order Date']
).dt.days

sales_df[['Order Date','Ship Date','Shipping Days']].head()
```

```python
sales_df['Year'] = sales_df['Order Date'].dt.year
sales_df['Month'] = sales_df['Order Date'].dt.month
sales_df['Month Name'] = sales_df['Order Date'].dt.month_name()

sales_df['Quarter'] = sales_df['Order Date'].dt.quarter

sales_df['Week Number'] = sales_df['Order Date'].dt.isocalendar().week.astype(int)

sales_df['Day of Week'] = sales_df['Order Date'].dt.day_name()

sales_df['Day'] = sales_df['Order Date'].dt.day

sales_df.head()
```

```python
def get_season(month):
    if month in [12, 1, 2]:
        return "Winter"
    elif month in [3, 4, 5]:
        return "Spring"
    elif month in [6, 7, 8]:
        return "Summer"
    else:
        return "Autumn"

sales_df["Season"] = sales_df["Month"].apply(get_season)

sales_df[['Order Date','Month','Season']].head()
```

```python
plt.figure(figsize=(14,6))

plt.plot(
    monthly_sales['Order Date'],
    monthly_sales['Sales'],
    linewidth=2,
    marker='o'
)

plt.title("Overall Monthly Sales Trend (2015–2018)", fontsize=16)
plt.xlabel("Order Date")
plt.ylabel("Total Monthly Sales")

plt.grid(True)

plt.tight_layout()

plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
monthly_ts = monthly_sales.set_index('Order Date')

# Decompose
decomposition = seasonal_decompose(
    monthly_ts['Sales'],
    model='additive',
    period=12
)
```

```python
fig = decomposition.plot()

fig.set_size_inches(15,10)

plt.tight_layout()

plt.show()
```

The time series decomposition separates the monthly sales into trend, seasonal, and residual components.

**Trend:** The trend component shows a gradual increase in sales over time, particularly during the later years, indicating long-term business growth.
**Seasonality:** A clear seasonal pattern is observed, with recurring peaks and declines occurring at similar periods each year. This suggests that customer demand follows a predictable annual cycle.
**Residual:** The residual component contains irregular fluctuations that the trend or seasonal effects cannot explain. These variations may represent unexpected market events, promotional campaigns, or other external business factors.

```python
from statsmodels.tsa.stattools import adfuller
```

```python
result = adfuller(monthly_ts['Sales'])

print("ADF Statistic :", result[0])
print("p-value :", result[1])
print("Number of Lags :", result[2])
print("Number of Observations :", result[3])

print("\nCritical Values:")

for key, value in result[4].items():
    print(f"{key}: {value}")
```

The Augmented Dickey-Fuller (ADF) test was performed to determine whether the monthly sales time series is stationary.

The obtained p-value was 0.000278, which is less than the significance level of 0.05. Therefore, the null hypothesis of non-stationarity is rejected, indicating that the monthly sales series is already stationary. Since the data satisfies the stationarity requirement, it is suitable for statistical forecasting models such as SARIMA without requiring mandatory differencing.

```python
if result[1] < 0.05:
    print("\nThe time series is stationary.")
else:
    print("\nThe time series is NOT stationary.")
```

```python
monthly_diff = monthly_ts['Sales'].diff().dropna()

plt.figure(figsize=(14,5))

plt.plot(monthly_diff)

plt.title("Monthly Sales After First-Order Differencing")
plt.xlabel("Order Date")
plt.ylabel("Differenced Sales")

plt.grid(True)

plt.show()
```

```python
result_diff = adfuller(monthly_diff)

print("ADF Statistic :", result_diff[0])
print("p-value :", result_diff[1])

if result_diff[1] < 0.05:
    print("\nThe differenced series is now stationary.")
else:
    print("\nThe differenced series is still not stationary.")
```

Although the original monthly sales series was already stationary, first-order differencing was performed as an additional demonstration. The differenced series also passed the Augmented Dickey-Fuller test with an even lower p-value, confirming strong stationarity. Since the original series was already stationary, the differencing step was not essential for this dataset but demonstrates the standard preprocessing technique commonly applied to non-stationary time series.

```python
train = monthly_ts.iloc[:-3]
test = monthly_ts.iloc[-3:]

print("Training Data Shape:", train.shape)
print("Testing Data Shape:", test.shape)

print("\nTraining Period:")
print(train.index.min(), "to", train.index.max())

print("\nTesting Period:")
print(test.index.min(), "to", test.index.max())
```

**Sales Forecasting Using Three Different Models**

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX

from sklearn.metrics import (
    mean_absolute_error,
    mean_squared_error,
    mean_absolute_percentage_error
)

import numpy as np
```

**Model 1: SARIMA (Statistical Forecasting)**

```python
sarima_model = SARIMAX(
    train['Sales'],
    order=(1,1,1),
    seasonal_order=(1,1,1,12),
    enforce_stationarity=False,
    enforce_invertibility=False
)

sarima_result = sarima_model.fit()

print(sarima_result.summary())
```

*The warning occurs because the dataset is relatively small for estimating a seasonal SARIMA model with multiple parameters. Despite the warning, the model converged successfully and produced valid forecasts.*

**Forecast on Test Data**

```python
sarima_forecast = sarima_result.get_forecast(
    steps=len(test)
)

forecast_mean = sarima_forecast.predicted_mean

confidence_interval = sarima_forecast.conf_int()

forecast_mean
```

**Plot Actual vs Forecast**

```python
plt.figure(figsize=(14,6))

plt.plot(
    train.index,
    train['Sales'],
    label='Training Data',
    linewidth=2
)

plt.plot(
    test.index,
    test['Sales'],
    label='Actual Sales',
    linewidth=2,
    marker='o'
)

plt.plot(
    forecast_mean.index,
    forecast_mean,
    label='SARIMA Forecast',
    linewidth=2,
    marker='o'
)

plt.fill_between(
    confidence_interval.index,
    confidence_interval.iloc[:,0],
    confidence_interval.iloc[:,1],
    alpha=0.3,
    label='95% Confidence Interval'
)

plt.title("SARIMA Forecast vs Actual")

plt.xlabel("Date")

plt.ylabel("Sales")

plt.legend()

plt.grid(True)

plt.show()
```

**Evaluate the Model**

```python
sarima_mae = mean_absolute_error(
    test['Sales'],
    forecast_mean
)

sarima_rmse = np.sqrt(
    mean_squared_error(
        test['Sales'],
        forecast_mean
    )
)

sarima_mape = mean_absolute_percentage_error(
    test['Sales'],
    forecast_mean
)

print(f"MAE  : {sarima_mae:.2f}")
print(f"RMSE : {sarima_rmse:.2f}")
print(f"MAPE : {sarima_mape:.4f}")
```

**Future Forecast**

```python
future_forecast = sarima_result.get_forecast(
    steps=3
)

future_sales = future_forecast.predicted_mean

future_ci = future_forecast.conf_int()

future_sales
```

**Plot Future Forecast**

```python
plt.figure(figsize=(14,6))

plt.plot(
    monthly_ts.index,
    monthly_ts['Sales'],
    label='Historical Sales',
    linewidth=2
)

plt.plot(
    future_sales.index,
    future_sales,
    marker='o',
    linewidth=2,
    label='Future Forecast'
)

plt.fill_between(
    future_ci.index,
    future_ci.iloc[:,0],
    future_ci.iloc[:,1],
    alpha=0.3
)

plt.title("3-Month SARIMA Sales Forecast")

plt.xlabel("Date")
plt.ylabel("Sales")

plt.legend()

plt.grid(True)

plt.show()
```

**SARIMA Model Results**

The SARIMA(1,1,1)(1,1,1,12) model was successfully trained using the monthly sales data. The model achieved an **AIC of 398.775**, which serves as a baseline for evaluating future parameter optimizations. The forecast closely follows the overall upward trend in sales and captures the general direction of demand during the testing period.

The model produced the following evaluation metrics:

- **Mean Absolute Error (MAE): 19,244.49**
- **Root Mean Squared Error (RMSE): 19,950.07**
- **Mean Absolute Percentage Error (MAPE): 20.53%**

The forecast indicates increasing sales during the final three months, although the model slightly underestimates the sharp sales peak observed in November 2018. The confidence interval widens for future predictions, reflecting increasing uncertainty as the forecasting horizon extends.

**Model 3: XGBoost for Time Series**

```python
from xgboost import XGBRegressor
```

**Feature Engineering**

```python
xgb_df = monthly_sales.copy()

# Date Features
xgb_df['Year'] = xgb_df['Order Date'].dt.year
xgb_df['Month'] = xgb_df['Order Date'].dt.month
xgb_df['Quarter'] = xgb_df['Order Date'].dt.quarter

# Season Feature
def get_season(month):
    if month in [12, 1, 2]:
        return "Winter"
    elif month in [3, 4, 5]:
        return "Spring"
    elif month in [6, 7, 8]:
        return "Summer"
    else:
        return "Autumn"

xgb_df['Season'] = xgb_df['Month'].apply(get_season)

# Lag Features
xgb_df['Lag_1'] = xgb_df['Sales'].shift(1)
xgb_df['Lag_2'] = xgb_df['Sales'].shift(2)
xgb_df['Lag_3'] = xgb_df['Sales'].shift(3)

# Rolling Mean
xgb_df['Rolling_Mean_3'] = (
    xgb_df['Sales']
    .rolling(3)
    .mean()
)

# Remove missing values
xgb_df.dropna(inplace=True)

# One-Hot Encode Season
xgb_df = pd.get_dummies(
    xgb_df,
    columns=['Season'],
    drop_first=True
)

display(xgb_df.head())

print(xgb_df.columns)
```

```python
features = [col for col in xgb_df.columns if col not in ['Order Date', 'Sales']]

X = xgb_df[features]
y = xgb_df['Sales']
```

**Remove Missing Rows**

```python
xgb_df = xgb_df.dropna().reset_index(drop=True)

print(xgb_df.shape)
```

**Prepare Features**

```python
features = [
    'Lag_1',
    'Lag_2',
    'Lag_3',
    'Rolling_Mean_3',
    'Month',
    'Quarter',
    'Year',
    'Season_Spring',
    'Season_Summer',
    'Season_Winter'
]

X = xgb_df[features]

y = xgb_df['Sales']
```

**Train-Test Split**

```python
X_train = X.iloc[:-3]
X_test = X.iloc[-3:]

y_train = y.iloc[:-3]
y_test = y.iloc[-3:]
```

**Train the Model**

```python
xgb_model = XGBRegressor(
    n_estimators=100,
    learning_rate=0.05,
    max_depth=3,
    random_state=42
)

xgb_model.fit(X_train, y_train)
```

**Predictions**

```python
xgb_predictions = xgb_model.predict(X_test)

xgb_predictions
```

**Evaluation**

```python
xgb_mae = mean_absolute_error(
    y_test,
    xgb_predictions
)

xgb_rmse = np.sqrt(
    mean_squared_error(
        y_test,
        xgb_predictions
    )
)

xgb_mape = mean_absolute_percentage_error(
    y_test,
    xgb_predictions
)

print(f"MAE  : {xgb_mae:.2f}")
print(f"RMSE : {xgb_rmse:.2f}")
print(f"MAPE : {xgb_mape:.4f}")
```

**Actual vs Predicted**

```python
plt.figure(figsize=(12,6))

plt.plot(
    y_test.values,
    marker='o',
    linewidth=2,
    label='Actual'
)

plt.plot(
    xgb_predictions,
    marker='o',
    linewidth=2,
    label='Predicted'
)

plt.title("XGBoost Forecast vs Actual")

plt.xlabel("Test Samples")

plt.ylabel("Sales")

plt.legend()

plt.grid(True)

plt.show()
```

**XGBoost Forecasting Results**

The XGBoost Regressor was trained using engineered time-series features, including lag variables, rolling mean, month, quarter, year, and season indicators. Unlike SARIMA and Prophet, XGBoost treats the forecasting task as a supervised machine learning problem by learning relationships between historical observations and future sales.

The model achieved the following performance on the testing dataset:

- **Mean Absolute Error (MAE): 14,690.36**
- **Root Mean Squared Error (RMSE): 20,506.42**
- **Mean Absolute Percentage Error (MAPE): 13.91%**

Among the three forecasting models, XGBoost produced the lowest MAE and MAPE, indicating more accurate overall predictions. Although SARIMA achieved a slightly lower RMSE, XGBoost demonstrated superior prediction accuracy in terms of average and percentage errors, making it the preferred model for forecasting monthly sales in this project.

**Task 4: Product Category & Region Level Forecasting**

```python
from xgboost import XGBRegressor

def forecast_with_xgboost(df, label):
    """
    Train an XGBoost model on a category/region monthly sales series
    and forecast the next 3 months.
    """

    temp = df.copy()

    # Monthly aggregation
    temp = (
        temp.groupby(pd.Grouper(key='Order Date', freq='ME'))['Sales']
        .sum()
        .reset_index()
    )

    # Time Features
    temp['Year'] = temp['Order Date'].dt.year
    temp['Month'] = temp['Order Date'].dt.month
    temp['Quarter'] = temp['Order Date'].dt.quarter

    # Lag Features
    temp['Lag_1'] = temp['Sales'].shift(1)
    temp['Lag_2'] = temp['Sales'].shift(2)
    temp['Lag_3'] = temp['Sales'].shift(3)

    # Rolling Mean
    temp['Rolling_Mean_3'] = (
        temp['Sales']
        .rolling(3)
        .mean()
    )

    temp.dropna(inplace=True)

    features = [
        'Lag_1',
        'Lag_2',
        'Lag_3',
        'Rolling_Mean_3',
        'Month',
        'Quarter',
        'Year'
    ]

    X = temp[features]
    y = temp['Sales']

    model = XGBRegressor(
        n_estimators=100,
        learning_rate=0.05,
        max_depth=3,
        random_state=42
    )

    model.fit(X, y)

    # ---------- Recursive Forecast ----------
    history = temp.copy()

    predictions = []

    for _ in range(3):

        next_month = history.iloc[-1]['Month'] + 1
        next_year = history.iloc[-1]['Year']

        if next_month == 13:
            next_month = 1
            next_year += 1

        next_quarter = (next_month - 1) // 3 + 1

        lag1 = history.iloc[-1]['Sales']
        lag2 = history.iloc[-2]['Sales']
        lag3 = history.iloc[-3]['Sales']

        rolling = history.iloc[-3:]['Sales'].mean()

        new_row = pd.DataFrame({
            'Lag_1':[lag1],
            'Lag_2':[lag2],
            'Lag_3':[lag3],
            'Rolling_Mean_3':[rolling],
            'Month':[next_month],
            'Quarter':[next_quarter],
            'Year':[next_year]
        })

        pred = model.predict(new_row)[0]

        predictions.append(pred)

        history.loc[len(history)] = [
            pd.NaT,
            pred,
            next_year,
            next_month,
            next_quarter,
            lag1,
            lag2,
            lag3,
            rolling
        ]

    return predictions
```

```python
from sklearn.ensemble import IsolationForest
```

**Train Isolation Forest**

```python
iso_forest = IsolationForest(
    contamination=0.08,
    random_state=42
)

weekly_sales['Isolation_Anomaly'] = iso_forest.fit_predict(
    weekly_sales[['Sales']]
)
```

**Create Anomaly Dataset**

```python
isolation_anomalies = weekly_sales[
    weekly_sales['Isolation_Anomaly'] == -1
]

print("Number of anomalies detected:",
      len(isolation_anomalies))

display(isolation_anomalies.head())
```

**Plot Isolation Forest Anomalies**

```python
plt.figure(figsize=(15,6))

# Weekly Sales
plt.plot(
    weekly_sales['Order Date'],
    weekly_sales['Sales'],
    linewidth=2,
    label='Weekly Sales'
)

# Highlight anomalies
plt.scatter(
    isolation_anomalies['Order Date'],
    isolation_anomalies['Sales'],
    color='red',
    s=80,
    label='Anomaly'
)

plt.title("Isolation Forest - Weekly Sales Anomalies")

plt.xlabel("Date")

plt.ylabel("Weekly Sales")

plt.legend()

plt.grid(True)

plt.show()
```

**Rolling Mean and Standard Deviation**

```python
weekly_sales['Rolling Mean'] = (
    weekly_sales['Sales']
    .rolling(window=4)
    .mean()
)

weekly_sales['Rolling Std'] = (
    weekly_sales['Sales']
    .rolling(window=4)
    .std()
)
```

**Calculate Z-Score**

```python
weekly_sales['Z_Score'] = (
    weekly_sales['Sales']
    - weekly_sales['Rolling Mean']
) / weekly_sales['Rolling Std']
```

**Detect Anomalies**

```python
weekly_sales['Z_Anomaly'] = (
    weekly_sales['Z_Score'].abs() > 2
)

zscore_anomalies = weekly_sales[
    weekly_sales['Z_Anomaly']
]

print("Number of Z-Score anomalies:",
      len(zscore_anomalies))
```

**Plot Z-Score Anomalies**

```python
plt.figure(figsize=(15,6))

plt.plot(
    weekly_sales['Order Date'],
    weekly_sales['Sales'],
    linewidth=2,
    label='Weekly Sales'
)

plt.scatter(
    zscore_anomalies['Order Date'],
    zscore_anomalies['Sales'],
    color='orange',
    s=80,
    label='Z-Score Anomaly'
)

plt.title("Rolling Z-Score Weekly Sales Anomalies")

plt.xlabel("Date")

plt.ylabel("Weekly Sales")

plt.legend()

plt.grid(True)

plt.show()
```

**Combined Anomaly Report**

```python
report = weekly_sales.copy()

report['Isolation Forest'] = report['Isolation_Anomaly'].map({
    1: 'Normal',
    -1: 'Anomaly'
})

report['Z-Score'] = report['Z_Anomaly'].map({
    True: 'Anomaly',
    False: 'Normal'
})

report['Possible Reason'] = "Normal"

for i in report.index:
    if report.loc[i, 'Isolation Forest'] == 'Anomaly':
        if report.loc[i, 'Sales'] > report['Sales'].median():
            report.loc[i, 'Possible Reason'] = "High sales spike (promotion or festive demand)"
        else:
            report.loc[i, 'Possible Reason'] = "Unusually low sales (supply or demand issue)"
    elif report.loc[i, 'Z-Score'] == 'Anomaly':
        report.loc[i, 'Possible Reason'] = "Large deviation from rolling average"

report[['Order Date', 'Sales', 'Isolation Forest', 'Z-Score', 'Possible Reason']]
```

Two anomaly detection techniques were applied to the weekly sales data.

The Isolation Forest algorithm successfully detected multiple anomalous weeks characterized by unusually high or low sales values. Several anomalies occurred during periods of exceptionally high sales, particularly toward the end of the dataset, while a few unusually low sales observations were also identified. These anomalies may correspond to promotional campaigns, seasonal shopping events, inventory clearance activities, or unexpected changes in customer demand.

The rolling Z-score method did not identify any observations exceeding the predefined threshold of two standard deviations. This indicates that, within the selected rolling window, sales variations remained within the expected statistical range.

Overall, Isolation Forest proved to be more effective for detecting anomalies in this retail sales dataset because it considers the overall distribution of the data rather than relying solely on local statistical deviations.

**Task 6: Product Demand Segmentation Using K-Means Clustering**

**Aggregate Product Sales**

```python
product_sales = (
    sales_df
    .groupby('Product Name')
    .agg({
        'Sales':'sum'
    })
    .reset_index()
)

product_sales.head()
```

```python
from sklearn.cluster import KMeans

from sklearn.preprocessing import StandardScaler
```

**Scale the Data**

```python
scaler = StandardScaler()

scaled_sales = scaler.fit_transform(
    product_sales[['Sales']]
)
```

**Elbow Method**

```python
wcss = []

for i in range(1,11):

    model = KMeans(
        n_clusters=i,
        random_state=42,
        n_init=10
    )

    model.fit(scaled_sales)

    wcss.append(model.inertia_)
```

**Plot Elbow Curve**

```python
plt.figure(figsize=(8,5))

plt.plot(
    range(1,11),
    wcss,
    marker='o',
    linewidth=2
)

plt.title("Elbow Method")

plt.xlabel("Number of Clusters")

plt.ylabel("WCSS")

plt.grid(True)

plt.show()
```

**Choosing the Number of Clusters**

The Elbow Method was used to determine the optimal number of clusters for K-Means clustering. The Within-Cluster Sum of Squares (WCSS) decreased rapidly up to **three clusters**, after which the rate of improvement became much smaller. Therefore, **three clusters** were selected to represent low-, medium-, and high-demand product groups.

**Train K-Means**

```python
kmeans = KMeans(
    n_clusters=3,
    random_state=42,
    n_init=10
)

product_sales['Cluster'] = kmeans.fit_predict(
    scaled_sales
)

product_sales.head()
```

**Rename Clusters**

```python
cluster_avg = (
    product_sales
    .groupby('Cluster')['Sales']
    .mean()
    .sort_values()
)

cluster_avg
```

```python
cluster_map = {
    0: "Low Demand",
    2: "Medium Demand",
    1: "High Demand"
}

product_sales['Demand Segment'] = (
    product_sales['Cluster']
    .map(cluster_map)
)
```

**Cluster Visualization**

```python
plt.figure(figsize=(10,6))

for segment in product_sales['Demand Segment'].unique():

    subset = product_sales[
        product_sales['Demand Segment'] == segment
    ]

    plt.scatter(
        subset.index,
        subset['Sales'],
        label=segment,
        s=50
    )

plt.title("Product Demand Segmentation")

plt.xlabel("Products")

plt.ylabel("Total Sales")

plt.legend()

plt.grid(True)

plt.show()
```

**Task 6 Observation**

Products were grouped into three demand segments using the K-Means clustering algorithm based on their total sales.

The clustering results identified three distinct groups representing low-, medium-, and high-demand products. Most products belong to the low-demand segment, indicating that a large portion of the product catalog contributes relatively small sales individually. A moderate number of products fall into the medium-demand segment, while only a small group of products generates exceptionally high sales.

These demand segments provide valuable business insights for inventory management, warehouse planning, and product prioritization. High-demand products should receive priority for stock replenishment and marketing, medium-demand products require balanced inventory management, and low-demand products may benefit from promotional strategies or inventory optimization.
